```python
train=pd.read_csv('../input/15cse380-nndl-eval/train.csv',na_values=[float('nan')])
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# train=train.fillna(method=)
```

```python
train['region']=le.fit_transform(train['region'])
# train.drop('region',inplace=True,axis=1)
# train['region'].corr(train['is_promoted'])
```

```python
train['department']=le.fit_transform(train['department'])
# train.drop('department',inplace=True,axis=1)
```

```python
# train['department'].corr(train['is_promoted'])
```

```python

train=train.join(pd.get_dummies(train.department), how='right', lsuffix='depl', rsuffix='depr')
train.drop('department',axis=1,inplace=True)
```

```python
train['education']=train['education'].fillna(train['education'].mode(), inplace=True)
```

```python
train['education']=le.fit_transform(train['education'])
```

```python
train=train.join(pd.get_dummies(train.education), how='right', lsuffix='edul', rsuffix='edur')
```

```python
train.drop('education',axis=1,inplace=True)
```

```python
train.head(5)
```

```python

train['gender']=le.fit_transform(train['gender'])
# train['gender'].corr(train['is_promoted'])
```

```python
train.corr()
```

```python
train.recruitment_channel=le.fit_transform(train.recruitment_channel)

train=train.join(pd.get_dummies(train.recruitment_channel), how='right', lsuffix='recruitl', rsuffix='recrutr')
train.drop('recruitment_channel',inplace=True,axis=1)
```

```python
train['previous_year_rating'].fillna(0,inplace=True)
train=train.join(pd.get_dummies(train.previous_year_rating), how='right', lsuffix='prevl', rsuffix='prevr')
train.drop('previous_year_rating',inplace=True,axis=1)
```

```python
train=train.join(pd.get_dummies(train.region), how='right', lsuffix='regl', rsuffix='regr')
train.drop('region',inplace=True,axis=1)
```

```python

train.columns
```

```python
train.drop('employee_id',axis=1,inplace=True)
c=train.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
c
```

```python
sc=MinMaxScaler()
```

```python
train['age']=sc.fit_transform(np.array(train['age']).reshape(-1,1))
```

```python
train['avg_training_score']=sc.fit_transform(np.array(train['avg_training_score']).reshape(-1,1))
```

```python
train['length_of_service']=sc.fit_transform(np.array(train['length_of_service']).reshape(-1,1))
```

```python
train.head(20)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,BatchNormalization,Dropout
```

```python
model=Sequential()
model.add(Dense(input_dim=len(c)-1,units=100,activation='relu'))

model.add(Dense(units=100,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=70,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=63,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(units=60,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.3))
# model.add(BatchNormalization())
model.add(Dense(30,activation='relu'))
# model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))

```

```python
from tensorflow.keras import optimizers
adm = optimizers.Adam(lr=0.0001)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```
